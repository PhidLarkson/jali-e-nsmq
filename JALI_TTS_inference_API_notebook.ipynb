```python
config_path1="/content/config.json"
model_path1="/content/coqui_vits.onnx"
config1 = VitsConfig()
config1.load_json(config_path1)
vits1 = Vits.init_from_config(config1)
vits1.load_onnx(model_path1)
```

```python
config_pat2="" #student path
model_path2="" #student path
config2 = VitsConfig()
config2.load_json(config_path2)
vits2 = Vits.init_from_config(config2)
vits2.load_onnx(model_path2)
```

```python
def tts_call(text:str):
  text_inputs = np.asarray(
      vits.tokenizer.text_to_ids(text, language="en"),
      dtype=np.int64,
  )[None, :]
  audio = vits.inference_onnx(text_inputs, speaker_id=0)
  with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as temp_file:
    out_path = temp_file.name
  save_wav(wav=audio[0], path=out_path,sample_rate=22050)
  return out_path
```

```python
app = FastAPI()
app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)
```

```python
class TextRequest(BaseModel):
    text: str

class OutputText(BaseModel):
  answer: str
```

```python
@app.get("/synthesize_audios")
def synthesize_audio(payload: TextRequest):
    # Call your TTS function to generate the audio file
    audio_file_path = tts_call(payload.text)
    # Return the audio file as the response
    return FileResponse(audio_file_path, media_type="audio/wav")

@app.get("/", response_model=OutputText)
async def root():
    return {"answer": "Hello from TTS"}
```
